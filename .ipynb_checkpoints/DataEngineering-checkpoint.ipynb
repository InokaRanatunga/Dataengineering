```python
#Wind Strength Categorization:
def categorize_wind_kmh(speed):
    if speed <= 5.4:
        return "Calm"
    elif speed <= 11.9:
        return "Light Air"
    elif speed <= 19.4:
        return "Light Breeze"
    elif speed <= 28.4:
        return "Gentle Breeze"
    elif speed <= 38.5:
        return "Moderate Breeze"
    elif speed <= 49.7:
        return "Fresh Breeze"
    elif speed <= 61.6:
        return "Strong Breeze"
    elif speed <= 74.5:
        return "Near Gale"
    elif speed <= 87.8:
        return "Gale"
    elif speed <= 102.2:
        return "Strong Gale"
    elif speed <= 117.4:
        return "Storm"
    else:
        return "Violent Storm"

df['wind_strength'] = df['Wind Speed (km/h)'].apply(categorize_wind)
```

```python

df1=df.copy()
df1 = df1.set_index("Formatted Date")


# ---------------------------------------------------------
# 4. DAILY AVERAGES (Calculate Mean per Day)
# ---------------------------------------------------------
daily_avg = df1.resample("D").agg({
    "Temperature (C)": "mean",
    "Humidity": "mean",
    "Wind Speed (km/h)": "mean"
}).reset_index()

daily_avg = daily_avg.rename(columns={
    "Temperature (C)": "avg_temperature_c",
    "Humidity": "avg_humidity",
    "Wind Speed (km/h)": "avg_wind_speed_kmh"
})

# ---------------------------------------------------------
# 5. DAILY RAW VALUES (First record of each day)
# ---------------------------------------------------------
daily_raw = df1.sort_index().groupby(pd.Grouper(freq='D')).first().reset_index()

# ---------------------------------------------------------
# 6. MERGE RAW DAILY DATA + DAILY AVERAGES
# ---------------------------------------------------------
daily_df = daily_raw.merge(
    daily_avg,
    on="Formatted Date",
    how="left"
)

# ---------------------------------------------------------
# 7. RENAME COLUMNS AT THE END
# ---------------------------------------------------------
daily_df = daily_df.rename(columns={
    "Formatted Date": "formatted_date",
    "Precip Type": "precip_type",
    "Temperature (C)": "temperature_c",
    "Apparent Temperature (C)": "apparent_temperature_c",
    "Humidity": "humidity",
    "Wind Speed (km/h)": "wind_speed_kmh",
    "Visibility (km)": "visibility_km",
    "Pressure (millibars)": "pressure_millibars"
})

# ---------------------------------------------------------
# 8. FINAL DAILY COLUMNS
# ---------------------------------------------------------
daily_df = daily_df[[
    "formatted_date",
    "precip_type",
    "temperature_c",
    "apparent_temperature_c",
    "humidity",
    "wind_speed_kmh",
    "visibility_km",
    "pressure_millibars",
    "wind_strength",
    "avg_temperature_c",
    "avg_humidity",
    "avg_wind_speed_kmh"
]]

print(daily_df.head())
```

```python
#Create MONTHLY aggregates
#Add year-month columns
df['year'] = df['Formatted Date'].dt.year
df['month'] = df['Formatted Date'].dt.month
```

```python
#Monthly averages
monthly_df = df.groupby(['year', 'month']).agg({
    'Temperature (C)': 'mean',
    'Apparent Temperature (C)': 'mean',
    'Humidity': 'mean',
    'Visibility (km)': 'mean',
    'Pressure (millibars)': 'mean'
}).reset_index()
```

```python
#Rename:
monthly_df.rename(columns={
    'Temperature (C)': 'avg_temperature_c',
    'Apparent Temperature (C)': 'avg_apparent_temperature_c',
    'Humidity': 'avg_humidity',
    'Visibility (km)': 'avg_visibility_km',
    'Pressure (millibars)': 'avg_pressure_millibars'
}, inplace=True)
```

```python
#Add MONTHLY MODE of “Precip Type”
mode_precip = (
    df.groupby(['year', 'month'])['Precip Type']
      .agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
      .reset_index()
      .rename(columns={'Precip Type': 'mode_precip_type'})
)
```

```python
#Merge into monthly_df:
monthly_df = monthly_df.merge(mode_precip, on=['year', 'month'], how='left')
```

```python
#Save daily file:
daily_df.to_csv("daily_weather.csv", index=False)
print("Saved: daily_weather.csv")
```

```python
#Save monthly file:
monthly_df.to_csv("monthly_weather.csv", index=False)
print("Saved: monthly_weather.csv")
```
